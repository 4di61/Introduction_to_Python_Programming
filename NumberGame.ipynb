Problem Statement

Geeta and Seeta are playing a number game, where for every iteration a random number (1 to 50) will be generated first for Geeta and then for Seeta. Input the number of iterations to be played from the user. They get points as follows:<br>
i.	For every even number, 2 point<br>
ii.	For every odd number, 3 points<br>
iii.	For every square number (even or odd), 4 points<br>
Print the points obtained by Geeta and Seeta


```python
import random
import math
```

```python

n= int(input("Enter the number of iterations: "))
# store the random numbers generated for each iteration for Geeta and Seeta in lists
g_list=[]
s_list=[]
# store the score for each iteration aciheved by Geeta and Seeta in lists
g_total=[]
s_total=[]

#Execute the for loop for the given number of iteraions
for i in range(n):
#Generate random numbers for Geeta and Seeta for each iteration
  g= random.randint(1,51)
  s= random.randint(1,51)
#Append the numbers generated to the individual list for Geeta and Seeta
  g_list.append(g)
  s_list.append(s)

#Check is the number generated by Geeta is a square number or even number or odd number. Append appropriate score in the list
  if math.sqrt(g) % 1==0:
    g_total.append(4)
  elif g%2==0:
    g_total.append(2)
  else:
    g_total.append(3)

#Check is the number generated by Seeta is a square number or even number or odd number. Append appropriate score in the list
  if math.sqrt(s) % 1==0:
    s_total.append(4)
  elif s%2==0:
    s_total.append(2)
  else:
    s_total.append(3)

#Find the total score for Geeta and Seeta
g_sum= sum(g_total)
s_sum= sum(s_total)

#Print all the random numbers generated for each iteration, score for each round and total score
print("Numbers generated for Geeta ", g_list)
print("Numbers generated for Seeta ", s_list)
print("Points for Geeta in each round ", g_total)
print("Points for Seeta in each round ", s_total)
print("Geeta's score is ",g_sum )
print("Seeta's score is ",s_sum )

#Declare the winner
if g_sum==s_sum:
  print("Its a draw!")
elif g_sum>s_sum:
  print("Winner is Geeta!")
else:
  print("Winner is Seeta!")   
```
